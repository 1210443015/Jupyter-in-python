```python
import pandas as pd
pd.set_option('display.max_row', 10)
```

```python
"""
对变量解释进行整理
"""
path = './data/mHisWellInfo.txt'

parameters_1 = []
parameters_2 = []
parameters_3 = []

with open(path, encoding='utf-8') as file:
    line = file.readline()
    parameters = line.split(',')
    for i in range(len(parameters)):
        if i % 2 == 0:
            split_list = parameters[i].split(':[')
            parameters_1.append(eval(split_list[0]))
            parameters_2.append(eval(split_list[1]))
        if i % 2 == 1:
            danwei = parameters[i].split(']')
            if danwei[0] == 'null':
                parameters_3.append(danwei[0])
            else:
                parameters_3.append(eval(danwei[0]))
                
df = pd.DataFrame({'变量': parameters_1, '变量解释': parameters_2, '变量单位': parameters_3})
df.to_csv('./data/parameters.csv', index=None)
```
